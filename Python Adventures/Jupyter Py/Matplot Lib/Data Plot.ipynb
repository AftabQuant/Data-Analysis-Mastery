```python
import matplotlib.pyplot as plt
x = ["A", "B", "C", "D", "E", "F"]
y = [20, 15, 10, 15, 20, 16]
plt.bar(x, y, edgecolor = "black")
plt.xlabel("Name")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
```

```python
x = ["A", "B", "C", "D", "E", "F"]
y = [20, 15, 10, 15, 20, 16]
z = [32, 20, 18, 10, 12, 2]
plt.plot(x, y, marker = ".", label = "Week1")
plt.plot(x, z, marker = ".", color = "green", label = "Week2")
plt.legend()
plt.show()
```

```python
import numpy as np
y = np.random.randint(1,10,50)
z = np.random.randint(1,100,50)
plt.scatter(y, z)
plt.show()
```

```python
x = ["A", "B", "C", "D", "E", "F"]
y = [20, 15, 10, 15, 20, 16]
s = [0, 0, 0, 0, 0, 0.1]
plt.pie(y, labels = x, explode = s)
plt.show()
```

```python
y = [20, 15, 10, 15, 20, 16, 34, 40, 32, 12. ]
plt.boxplot(y)
plt.show()
```

```python
y = np.random.randint(1,20,10)
plt.hist(y)
plt.show()
```

```python
y = np.random.randint(1,30,100)
plt.violinplot(y, showmeans = True)
plt.show()
```

```python
y = np.random.randint(1,20,10)
plt.stem(y)
plt.show()
```

```python
y = np.random.randint(1,20,10)
z = np.random.randint(1,20,10)
plt.stackplot(y, z)
plt.show()
```
