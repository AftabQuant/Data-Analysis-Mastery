```python
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    "Day" : [1,2,3,4,5],
    "nop" : [50,60,40,45,44]
}
sns.lineplot(data = data, x = "Day", y = "nop")
plt.show()
```

```python
data = sns.load_dataset("tips")
print(data)
print(data.info())
```

```python
sns.lineplot(data = data, x = "smoker", y = "total_bill", hue = "sex", palette = "viridis")
plt.show()
```

```python
sns.set_style(style = "darkgrid") # style 
sns.barplot(data = data, x = "day", y = "tip", hue = "sex", palette = "viridis")
plt.plot()
```

```python
sns.histplot(data = data, x = "day", hue = "smoker", kde = True)
plt.show()
```

```python
sns.scatterplot(data = data, x = "total_bill", y = "tip", hue = "smoker")
plt.show()
```

```python
gp = data.groupby("day").agg({"total_bill":"mean"})
sns.heatmap(gp)
plt.show()
```

```python
sns.countplot(data = data,x = "smoker")
plt.show()
```

```python
sns.violinplot(data=data, x = "total_bill")
plt.show()
```

```python
sns.pairplot(data, hue = "sex", palette = "viridis")
plt.show()
```

```python
sns.boxplot(data = data, palette = "viridis")
plt.show()
```

```python
# cat plot : categorical plot
sns.catplot(data= data, x = "tip", hue = "sex")
plt.show()
```

```python
sns.palplot(sns.color_palette("viridis"))
sns.palplot(sns.color_palette("viridis", 5))
```
