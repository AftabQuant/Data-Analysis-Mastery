```python
import pandas as pd
df = pd.read_excel("Amazon.xlsx")
#print(df.info())
df.loc[(df["quantity"] == 1), "Profit"] = "Yes"
df.loc[(df["quantity"] > 1), "Profit"] = "No"
print(df.head(15))
```

```python
df["Full Name"] = df["buyer"].str.lower() + " " + df["ship_city"]
print(df.head(15))
```

```python
df["Tax"] = df["quantity"] * 1/100
print(df.head(15))
```
